# Data Transformation using PySpark

**Lets add a column to TEAM_LEADER as using dbutils.Widgets**

```sql
%python

dbutils.widgets.text("Team_Leader", "Enter the name")
```

```sql
%python
name = dbutils.widgets.get("Team_Leader")
print(name)
```

```sql
%python
from pyspark.sql.functions import lit
dell_emp.withColumn('Team_Leader', lit(list(name))).display()
```
