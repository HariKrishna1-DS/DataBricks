#Another way of creating delta table using Dataframe API

Dataframe O/p Save as a table


```python
df = spark.read.format("csv").option("header","true").option("inferSchema","true").load("/Volumes/dev/haridb/myvolume/emp.csv")
df.display()
```

```python
df.write.format("delta").mode("append").saveAsTable("dev.delta_demodb.emps2")
```

```python
%sql
create table if not exists dev.delta_demodb.emp3 using delta as select * from dev.delta_demodb.emps2;

select * from dev.delta_demodb.emp3
```

```python
spark.sql("use dev.delta_demodb")
```

```python
spark.sql("show tables").display()
```

```python
spark.sql("select * from emp3").display()
```
