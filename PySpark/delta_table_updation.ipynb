# Method-2:
# Pyspark Style----->using Table_instance

### PySpark లో Table Instance అంటే, Databricks metastore లో register అయ్యి, cloud storage లో ఉన్న data ని represent చేసే permanent table object.

```python
from pyspark.sql.functions import *
from delta.tables import * 

deltatable_instance = DeltaTable.forName(spark,"dev1.demodb.emp")
deltatable_instance.update(condition = "ename = 'James'" , set = {"sal" : "5400"})
spark.sql("select * from dev1.demodb.emp").show()

#update dno
deltatable_instance.update(condition = "ename = 'Sony'" , set = {"dno" : "14"})
spark.sql("select * from dev1.demodb.emp").show()

#Update Name
deltatable_instance.update(condition = "eid = '104'" , set = {"ename": lit("Hari")})
spark.sql("select * from dev1.demodb.emp").show()
```
