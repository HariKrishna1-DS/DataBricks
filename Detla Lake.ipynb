# External Delta Tables
**An External Delta Table is a table where:**
Databricks manages only metadata
Data files are stored externally (you manage them)

```python
%sql
create table my_schema.external_data
(
  id int,
  name string,
  age int
)using delta
location 'adfss://sink@datalakeadb1/dfs.core.window.azure.net/my_schema/external_data'
```
