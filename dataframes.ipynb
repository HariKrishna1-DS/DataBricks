```python
df = spark.read.format("csv").option("header", "true").load("/Volumes/workspace/default/emp/emp1.csv")
print(df)
```

```python
df.display()
```

# 1.Select() 
** particular columns**

```python
df.select("Name", "salary").display()
```

# 2. Transformation
** Adding 5000 to all salary **

```python
df2 = df.select(df.Name,(df.salary+5000).alias("salary"))
display(df2)
```

#**OR** 

# 3. using SelectExpr() 

```python
df.selectExpr("Name","salary+5000 as salary").display()
```

```python
mydf = df
df.show()
```

# 4. Filter()
** I want those emps whose salary is greater than 20000**


```python
mydf.filter(mydf.salary>20000).show()
```

```python
#Filter ,select and show

mydf.filter(mydf.salary>20000).select('Name','salary').show()
```
